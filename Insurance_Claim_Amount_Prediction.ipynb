```python
# splitting the data into independent variable and dependent variable
x = df.drop(columns=['policy_id', 'claim_amount'], axis=1)
x.shape
```

```python
# Define column transformer for preprocessing
numerical_feature = x.select_dtypes(include=['int64','float64']).columns
categorical_feature = x.select_dtypes(include=['object']).columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer,KNNImputer
from sklearn.preprocessing import RobustScaler, StandardScaler, MinMaxScaler, OneHotEncoder,  LabelEncoder, OrdinalEncoder,FunctionTransformer
```

```python
median_imputer = SimpleImputer(strategy='median')
mode_imputer = SimpleImputer(strategy='most_frequent')
```

```python
robust_scaler = RobustScaler()
standard_scaler = StandardScaler()
min_max_scaler = MinMaxScaler()
```

```python
one_hot_encoder = OneHotEncoder(sparse_output=False)
label_encoder = LabelEncoder()
ordinal_encoder = OrdinalEncoder()
```

```python
cat_pipeline=Pipeline([("imputer",mode_imputer),("encoder",one_hot_encoder)])
num_pipeline=Pipeline([("imputer",median_imputer),("scaler",robust_scaler)])
```

```python
preprocessor_trans = ColumnTransformer([('num_pipeline', num_pipeline, numerical_feature),
                                               ('cat_pipeline', cat_pipeline, categorical_feature)], remainder='drop')
preprocessor_trans
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor, RandomForestRegressor, VotingRegressor, StackingRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr_model = Pipeline(steps=[("preprocessor", preprocessor_trans), ("Regressor", LinearRegression())])
ada_model = Pipeline(steps=[("preprocessor",preprocessor_trans), ("Regressor", AdaBoostRegressor())])
gra_model = Pipeline(steps=[("preprocessor", preprocessor_trans), ("Regressor", GradientBoostingRegressor())])
rf_model = Pipeline(steps=[("preprocessor", preprocessor_trans), ("Regressor", RandomForestRegressor())])
dt_model = Pipeline(steps=[("preprocessor", preprocessor_trans), ("Regressor", DecisionTreeRegressor())])

estimators = [('lr', lr_model), ('rf', rf_model),('gd',gra_model),('ada', ada_model), ('dt', dt_model)]
```

```python
stack_model = StackingRegressor(estimators= estimators, final_estimator= Ridge())
stack_model.fit(x_train, y_train)
```

```python
y_pred_test= stack_model.predict(x_test)
y_pred_train= stack_model.predict(x_train)
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)
r2_train = r2_score(y_train, y_pred_train)
r2_test = r2_score(y_test, y_pred_test)
print("StackingRegressor")
print(f"Mean Squared Erorr_train: {mse_train}, Mean Squared Erorr_test: {mse_test}")
print(f"R^2 SCORE_Train : {r2_train},R^2 SCORE_Test : {r2_test}")
```
